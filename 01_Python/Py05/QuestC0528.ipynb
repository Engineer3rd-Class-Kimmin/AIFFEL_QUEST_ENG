```python
#   Kiosk 클래스     
class Kiosk: 
    
    #   클래스 속성
    menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
    price = [2000, 3000, 3000, 2500, 2500, 3000]
    
    #   클래스 변수 설정시 초기값 세팅
    def __init__(self):
      
        self.menu = Kiosk.menu      #   메뉴
        self.price = Kiosk.price    #   가격
        self.temp = []              #   온도 리스트
        self.order_menu = []        #   주문 리스트
        self.order_price = []       #   가격 리스트

    #   메뉴 출력 
    def menu_print(self):
        
        #   메뉴와 가격을 for 문으로 출력
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')
            
        #   메뉴 출력 이후 즉시 주문 함수 호출
        self.menu_select()
        
    #   주문 
    def menu_select(self):
        
        #   메뉴 출력 한줄 띄우기
        print()  
        
        #   입력받을 메뉴 번호 초기화
        n = 0  
        
        # 주문 번호가 메뉴번호안에 있을때만 주문 이어가기
        while n < 1 or len(self.menu) < n:
            
            #   음료 번호 입력
            n = int(input("음료 번호를 입력하세요 : \n"))

            #   메뉴판에 있는 음료 번호일때
            if 1 <= n and n <= len(self.menu):
                
                #   선택 음료의 가격
                menu_price = self.price[n-1]  
                
                #   메뉴리스트에 선택 메뉴 넣기
                self.order_menu.append(self.menu[n-1])
                
                #   오더가격 리스트에 선택메뉴의 가격 넣기
                self.order_price.append(self.price[n-1])
                
                #   메뉴 선택되면 while 문 종료
                break
                
            # 메뉴판에 없는 번호일 때
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.\n")

                #   메뉴 선택으로 이동
                continue
            
        #   온도 선택 변수 초기화
        t = 0 
        
        #   1이나 2를 입력할 때까지 물어보기
        while t != 1 and t != 2:  
            
            #   온도 선택 입력 받기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : \n"))

            if t ==  1 :
                
                #   온도 리스트에 선택 온도 추가
                self.temp.append("HOT") 
                
                #   현재 선택한 온도/메뉴/가격 출력
                print('주문 음료 : HOT', self.menu[n-1], ' => ', menu_price , '원\n')

                #   추가 주문 함수 호출
                self.add_select()
                
                #   while 문 종료
                break
                
            elif t == 2 :
                
                #   온도 리스트에 선택 온도 추가
                self.temp.append("ICE") 
                
                #   현재 선택한 온도/메뉴/가격 출력
                print('주문 음료 : ICE', self.menu[n-1], ' => ', menu_price , '원\n')

                #   추가 주문 함수 호출
                self.add_select()
                
                #   while 문 종료
                break
                
            else:
                
                #   예외처리    
                print("1과 2 중 하나를 입력하세요.\n")
                
                #   온도 선택으로 이동
                continue
            
    #   추가 주문 함수        
    def add_select(self) : 
             
            #   추가 주문  선택하기              
            add_order = str(input("추가 주문 : y / n \n"))
        
            if add_order == "y" :
                
                #   메뉴 주문 함수 호출
                self.menu_select()
                
            #   n 선택시 주문 종료    
            elif add_order == "n" :       
                print("주문을 종료합니다")  
                
                #   오더 리스트 함수 호출      
                self.order_list()
                
            else :
                print("잘못된 입력입니다\n")
                       
                #   현재 함수 재 호출
                self.add_select()
        
      
    
    #   오더 리스트 함수            
    def order_list(self):  
        
        #   현재까지 주문한 메뉴의 총 가격
        a = sum(self.order_price) 
        
        #   주문표 만들기 위한 print
        print('⟝' + '-' * 30 + '⟞')
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        
        #   전체 주문 정보와 총 주문 금액 출력
        for x,y, in zip(self.temp,self.order_menu) :          
            print('상품명 : ',x,y) 
        
        #   주문표 만들기 위한 print    
        print('총 주문 금액 : ',a,"원")
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

        while True :
            
            #   주문 확인 후 결제 확인 할것
            b = str(input("주문 결제 진행하시겠습니까? y / n \n"))
          
            if b == "y" :
                
                #   지불 함수 호출
                self.pay()
                
                #   while 문은 탈출
                break
            
            #   주문 취소하면 프로그램을 종료함
            elif b == "n" :
                print("주문을 취소합니다")
                
                #   while 문 종료 및 프로그램 종료
                break
            
            else :
                print("잘못된 입력입니다\n")

                #   주문 결제 의사 재 확인
                continue
            
        return        
        
    #   지불 함수
    def pay(self) :
        
        #   결제 방법 선택
        a = int(input("결제 방법을 선택하세요 / 1은 현금 2는 카드 \n"))
    
        if a == 1 :
            
            print("직원을 호출하겠습니다")
            
            #   프로그램 종료
            return
        
        elif a == 2 :
            
            print("IC칩 방향에 맞게 카드를 꽂아주세요")
            
            #   프로그램 종료
            return
        
        else :
            print("잘못된 입력입니다\n")
          
            #   지불 함수 재호출
            self.pay()

#   변수에 Kiosk 클래스 할당  
a = Kiosk()

#   Kiosk의 menu_print 메서드 호출
a.menu_print()
```

```python
#   Square 클래스 
class Square:
    
    def __init__(self):
        
        #   계산하고 싶은 사각형 선택,0은 종료
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 0.종료 \n >>>'))

        if self.square == 1:
            
            #   직사각형 계산 함수 호출
            self.rect()

        elif self.square == 2:
            
            #   평행사변형 계산 함수 호출
            self.par()
        
        elif self.square == 3:
            
            #   사다리꼴 계산 함수 호출
            self.trape()
        
        elif self.square == 0 :
            print("프로그램을 종료합니다")
            
            #   프로그램 종료
            return
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')
            
            #   Square 클래스 재호출
            Square()
    
    #   직사각형 계산 함수        
    def rect(self):
        
        #   직사각형의 가로 세로 입력
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        
        #   넓이 계산후 변수에 저장
        area = width * vertical
        
        #   결과 출력
        print('직사각형의 넓이는 : ', area)
        
        #   한줄 띄우기
        print()
        
        #   Square class 재호출
        Square()      

    #   평행사변형 계산 함수
    def par(self):
        
        #   평행사변형의 가로 세로 입력
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        
        #   넓이 계산후 변수에 저장
        area = width * vertical
        
        #   결과 출력
        print('평행사변형의 넓이는 : ', area)
        
        #   한줄 띄우기
        print()
        
        #   Square class 재호출
        Square()

    #   사다리꼴 계산 함수
    def trape(self):
        
        #   사다리꼴의 아래변 윗변 높이 입력
        lowwidth, highwidth, vertical, = map(int, input('아래변,윗변,높이를 입력하세요. 예시 : 아래변,윗변,높이\n >>>').split(','))
        
        #   넓이 계산후 변수에 저장
        area = (lowwidth + highwidth) * vertical * 0.5
        
        #   결과 출력
        print('사다리꼴의 넓이는 : ', area)
        
        #   한줄 띄우기
        print()
        
        #   Square class 재호출
        Square() 

#   변수에 Square 클래스 할당 후 Square 클래스 호출
a = Square()  
```
