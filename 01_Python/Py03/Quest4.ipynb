```python
#   math 모듈을 import 하여 **연산시 사용할것
import math 

#   number1 초기화
number1 = 0
#   number2 초기화
number2 = 0
  
while True :
        
    try :
        
        #   첫번째 숫자 받는곳
        number1 = int(input("첫번째 숫자를 입력하세요")) 
        #   두번째 숫자 받는곳
        number2 = int(input("두번째 숫자를 입력하세요"))    
        #   연산자 입력
        operator = input("연산자를 입력하세요")
                
        #   연산자가 "+" 이면 + 계산값 출력
        if operator == "+" :
            number3 = number1 + number2
            print(number3)
                
            #   계산이 끝난 후 다시 계산할지 질문 "y" = 처음부터 다시 "n" = 프로그램 종료
            answer = str(input("계산을 계속 하시겠습니까 y / n "))
            if answer == "y" :
                continue
            else :
                print("계산기 프로그램을 종료합니다")
                break
                
        #   연산자가 "-" 이면 - 계산값 출력    
        elif operator == "-" :
            number3 = number1 - number2
            print(number3)
                
            #   계산이 끝난 후 다시 계산할지 질문 "y" = 처음부터 다시 "n" = 프로그램 종료
            answer = str(input("계산을 계속 하시겠습니까 y / n "))
            if answer == "y" :
                continue
            else :
                print("계산기 프로그램을 종료합니다")
                break
                
        #   연산자가 "/" 이면 / 계산값 출력   
        elif operator == "/" :
            number3 = number1 / number2
            print(number3)
                
            #   계산이 끝난 후 다시 계산할지 질문 "y" = 처음부터 다시 "n" = 프로그램 종료
            answer = str(input("계산을 계속 하시겠습니까 y / n "))
            if answer == "y" :
                continue
            else :
                print("계산기 프로그램을 종료합니다")
                break
                
        #   연산자가 "*" 이면 * 계산값 출력  
        elif operator == "*" :
            number3 = number1 * number2
            print(number3)
                
            #   계산이 끝난 후 다시 계산할지 질문 "y" = 처음부터 다시 "n" = 프로그램 종료
            answer = str(input("계산을 계속 하시겠습니까 y / n "))
            if answer == "y" :
                continue
            else :
                print("계산기 프로그램을 종료합니다")
                break
                
        #   연산자가 "**" 이면 math모듈 import 하여 ** 계산값 출력 
        elif operator == "**" :
            number3 = math.pow(number1,number2)
            print(number3)
                
            #   계산이 끝난 후 다시 계산할지 질문 "y" = 처음부터 다시 "n" = 프로그램 종료
            answer = str(input("계산을 계속 하시겠습니까 y / n "))
            if answer == "y" :
                continue
            else :
                print("계산기 프로그램을 종료합니다")
                break

    #   number2가 0 입력 시 나눗셈에서 0으로 나눌수 없음을 알림 
    except ZeroDivisionError :
        print("0 으로 나눌 수 없습니다")
    #   숫자 또는 연산자 입력 란에 올바른 입력이 아님을 알림
    except ValueError :
        print("정수가 아닙니다") 
        
#   while 문 안으로 try except 문을 넣어 에러가 나도 
#   다시 while문 위로 돌아가서 실행되게끔 코딩했다
#   코딩 경험이 없어서 내가 아는 코드 작성법으로 작성하다보니
#   쓸데없는 반복 코드가 많아지고 코드 전체가 지저분해졌다
```
