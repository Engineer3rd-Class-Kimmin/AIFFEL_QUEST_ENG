```python
import random


# Account 클래스 생성
class Account:

    # 클래스 변수
    account_count = 0

    # 생성자
    def __init__(self, name, money):

        # 은행 이름
        self.bank_name = "SC은행"

        # 예금주
        self.user = name

        # 잔액
        self.money = money

        # 입금 횟수
        self.deposit_count = 0

        # 입금 내역 리스트
        self.deposit_list = []

        # 출금 내역 리스트
        self.withdraw_list = []

        # 계좌번호 랜덤 생성
        front = random.randint(100, 999)
        middle = random.randint(10, 99)
        back = random.randint(100000, 999999)

        self.account_number = f"{front}-{middle}-{back}"

        # 계좌 개수 증가
        Account.account_count += 1

    # 계좌 개수 확인
    def get_account_num(self):

        print("현재 개설된 계좌 수 :",
              Account.account_count)

    # 입금 함수
    def deposit(self, price):

        # 1원 이상 입금 가능
        if price < 1:

            print("1원 이상 입금 가능합니다")
            return

        # 잔액 증가
        self.money += price

        # 입금 횟수 증가
        self.deposit_count += 1

        # 입금 내역 저장
        self.deposit_list.append(price)

        print(price, "원 입금 완료")
        print("현재 잔액 :", format(self.money, ","), "원")

        # 입금 5회마다 이자 지급
        if self.deposit_count % 5 == 0:

            interest = self.money * 0.01

            self.money += interest

            print("이자 지급 완료 :", format(int(interest), ","), "원")
            print("현재 잔액 :", format(int(self.money), ","), "원")

        print()

    # 출금 함수
    def withdraw(self, price):

        # 잔액 부족 체크
        if price > self.money:

            print("잔액 부족")
            return

        # 출금
        self.money -= price

        # 출금 내역 저장
        self.withdraw_list.append(price)

        print(price, "원 출금 완료")
        print("현재 잔액 :", format(self.money, ","), "원")
        print()

    # 계좌 정보 출력
    def display_info(self):

        print("은행이름 :", self.bank_name)
        print("예금주 :", self.user)
        print("계좌번호 :", self.account_number)
        print("잔고 :", format(int(self.money), ","), "원")

    # 입금 내역 출력
    def deposit_history(self):

        print("===== 입금 내역 =====")

        for i in self.deposit_list:

            print(format(i, ","), "원")

        print()

    # 출금 내역 출력
    def withdraw_history(self):

        print("===== 출금 내역 =====")

        for i in self.withdraw_list:

            print(format(i, ","), "원")

        print()


# 객체 생성
a1 = Account("kim", 1000000)
a2 = Account("lee", 50000)
a3 = Account("park", 3000000)

# 리스트 저장
account_list = [a1, a2, a3]


# 입금 
a1.deposit(5000)
a1.deposit(2000)
a1.deposit(3000)
a1.deposit(4000)
a1.deposit(5000)

# 출금 
a1.withdraw(10000)

# 계좌 정보 출력
a1.display_info()

print()

# 입금 내역 출력
a1.deposit_history()

# 출금 내역 출력
a1.withdraw_history()

# 계좌 개수 출력
a1.get_account_num()

print()

# 잔액 100만원 이상 고객 출력
print("===== 100만원 이상 고객 =====")

for i in account_list:

    if i.money >= 1000000:

        i.display_info()
        print()

#   객체를 만들고 그 객체를 다른 객체에 전달하여
#   다른 데이터를 생성하는 과정이 힘들었다
#   좀 더 유연하게 데이터 전달 분석 전처리 하는법을 배워야겠다
```

```python

# 랜덤 모듈 import
import random


# Character 클래스 생성
class Character:

    # 생성자
    def __init__(self, name, level, hp, attack, defence):
        
        #   캐릭터 이름
        self.name = name
 
        #   캐릭터 레벨
        self.level = level
        
        #   캐릭터 체력
        self.hp = hp
        
        #   캐릭터 공격력
        self.attack = attack

        #   캐릭터 방어력
        self.defence = defence

    # 살아있는지 확인
    def is_alive(self):

        if self.hp > 0:
            return True

        else:
            return False

    # 데미지 받기
    def take_damage(self, damage):

        # 방어력보다 공격력이 높을 때만 피해
        if damage > self.defence:

            #   입은 데미지 만큼 캐릭터 체력 감소
            self.hp -= (damage - self.defence)

        # 체력이 0 밑으로 내려가지 않게
        if self.hp < 0:
            self.hp = 0
            print(self.name , "님이 사망했습니다")

    # 공격하기
    def attack_target(self, target):

        # 1 ~ self.attack 사이 랜덤 데미지
        damage = random.randint(1, self.attack)

        print(self.name, "의 공격!")
        print("공격 데미지 :", damage)

        # 상대에게 데미지 전달
        target.take_damage(damage)

        print(self.name, "이(가)",
              target.name, "을 공격했습니다")


# Player 클래스 생성
class Player(Character):

    # Character 상속
    def __init__(self, name):

        # 부모 클래스 생성자 호출
        super().__init__(name, 1, 100, 25, 5)

        # 경험치 추가
        self.exp = 0

    # 경험치 획득
    def gain_experience(self, exp):

        self.exp += exp

        print(self.name, "경험치",
              exp, "획득")

    # 레벨업
    def level_up(self):

        # 경험치 50 이상이면
        if self.exp >= 50:

            self.level += 1
            self.attack += 10
            self.defence += 5

            # 경험치 차감
            self.exp -= 50

            print("레벨업!")
            print("현재 레벨 :", self.level)
            print("공격력 :", self.attack)
            print("방어력 :", self.defence)


# Monster 클래스 생성
class Monster(Character):

    # Character 상속
    def __init__(self, name, level):

        # 몬스터 능력치 랜덤 생성
        hp = random.randint(10, 30) * level

        attack = random.randint(5, 20) * level

        defence = random.randint(1, 5) * level

        # 부모 클래스 생성자 호출
        super().__init__(
            name,
            level,
            hp,
            attack,
            defence
        )


# 플레이어 생성
player = Player("용사")

# 몬스터 생성
monster = Monster("슬라임", 2)

# 공격 
player.attack_target(monster)

print("몬스터 현재 체력 :", monster.hp)

# 경험치 획득
player.gain_experience(50)

# 레벨업
player.level_up()



# 전투 함수
def battle(player, monster):

    # 둘 중 하나가 죽을 때까지 반복
    while player.is_alive() and monster.is_alive():

        # 플레이어 공격
        player.attack_target(monster)

        print(monster.name, "체력 :", monster.hp)

        # 몬스터가 죽었는지 확인
        if not monster.is_alive():
            break

        # 몬스터 공격
        monster.attack_target(player)

        print(player.name, "체력 :", player.hp)

        print("----------------")

    # 플레이어 승리
    if player.is_alive():

        exp = monster.level * 20

        player.gain_experience(exp)

        player.level_up()

        print("전투 승리!")

    # 플레이어 패배
    else:

        print("전투 패배..")
        
# 플레이어 생성
player = Player("용사")

# 몬스터 생성
monster = Monster("슬라임", 2)

# 전투 시작
battle(player, monster)


# 메인 함수
def main():

    # 몬스터 딕셔너리
    monster_dict = {
        "슬라임": 1,
        "고블린": 2,
        "오크": 3
    }

    # 사용자 이름 입력
    player_name = "모두연"

    # 플레이어 생성
    player = Player(player_name)

    print("게임 시작!")
    print("----------------")

    # 몬스터 순서대로 전투
    for monster_name, monster_level in monster_dict.items():

        print()
        print(monster_name, "등장!")
        print("레벨 :", monster_level)

        # 몬스터 생성
        monster = Monster(monster_name, monster_level)

        # 전투
        battle(player, monster)

        print("----------------")

        # 플레이어 사망 확인
        if not player.is_alive():

            print("게임 오버")
            return

    print("모든 몬스터 처치 완료!")
    
main()

#   클래스간의 연결과 몬스터,캐릭터를 구분하고
#   공격 또는 데미지를 받는 부분의 구현이 생각보다 어려웠다
#   연습을 더 많이 해봐야 하겠다
```
